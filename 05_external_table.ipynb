### creating the external location at gold level 

```python
%sql
CREATE EXTERNAL LOCATION if not exists gold_location
URL 'abfss://logdata@psgstorageacc.dfs.core.windows.net/gold/'
WITH (STORAGE CREDENTIAL acc_connector);
```

### # writing fact_log_detail(Union of both post & comment) to external location in delta format

```python
%sql
create external table if not exists de_project.default.fact_log_detail
using delta 
location 'abfss://logdata@psgstorageacc.dfs.core.windows.net/gold/log_detail'
```

```python
spark.table("de_project.gold.fact_log_detail").write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("de_project.default.fact_log_detail")
```
