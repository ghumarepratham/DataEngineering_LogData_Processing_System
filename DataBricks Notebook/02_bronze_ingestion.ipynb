```python
from pyspark.sql.functions import *

posts_df = spark.read.format("parquet") \
               .load("abfss://logdata@psgstorageacc.dfs.core.windows.net/bronze/postdata")

comments_df = spark.read.format("parquet") \
               .load("abfss://logdata@psgstorageacc.dfs.core.windows.net/bronze/comment/commentData")
```

```python
display(posts_df)
```

```python
display(comments_df)
```

```python
from pyspark.sql.functions import from_utc_timestamp, current_timestamp

posts_df = posts_df.withColumn("ingest_time", from_utc_timestamp(current_timestamp(), "Asia/Kolkata"))
comments_df = comments_df.withColumn("ingest_time", from_utc_timestamp(current_timestamp(), "Asia/Kolkata"))
```

### Checking the schema of data

```python
posts_df.printSchema()
```

```python
comments_df.printSchema()
```

### storing the raw data in bronze Schema

```python
posts_df.write.format("delta").mode("overwrite").option("mergeSchema", "true").saveAsTable("de_project.bronze.raw_posts")
comments_df.write.format("delta").mode("overwrite").option("mergeSchema", "true").saveAsTable("de_project.bronze.raw_comments")
```
